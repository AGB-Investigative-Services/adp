# Employee Enrollment Comparison Report - Employee Coverage Status - Enrollment Comparison

```python
import pandas as pd
import os
import numpy as np
import json

# Reading CSV

data = pd.read_csv('resources/data/Employee_Enrollments_Comparison.csv', encoding='ISO-8859-1')

# Removing 'Old' from the end of string
old = data['SELECTED DATE'].str.rstrip(' - Old')

# Removing 'New' from the end of string
new = old.str.rstrip(' - New')

# Adding new column
data['year'] = new

# Changing 'No Change' to 'Kept Coverage' and 'Add' to 'Added Coverage'
data['TYPE OF ENROLLMENT CHANGE'] = data['TYPE OF ENROLLMENT CHANGE'].replace('No Change', 'Kept Coverage')
data['TYPE OF ENROLLMENT CHANGE'] = data['TYPE OF ENROLLMENT CHANGE'].replace('Add', 'Added Coverage')

# Adding count column for value counts in groupbys
data['# of Employees'] = 1

# formatting date to datetime
data['year'] = data['year'].str[6:]

data = data.groupby(['year', 'TYPE OF ENROLLMENT CHANGE', 'PLAN TYPE', 'PLAN NAME', 'NAME']).sum()
data.reset_index(inplace=True)
data


data_json = data.to_json('resources/data/enrollment_comparsion_data.json', orient="records")
data_json = data.to_json(orient="records")
parsed = json.loads(data_json)
print(json.dumps(parsed, indent=4))
```
